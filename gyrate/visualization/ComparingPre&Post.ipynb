```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# --- Load data ---
scores_df = pd.read_csv("../data_cleaning/PreIRTscore.csv")
scores_df2 = pd.read_csv("../data_cleaning/PreIRTscoreExtended.csv")

theta_df = pd.read_csv("../fit_IRT/IRTMatrix_2PL_abilities.csv")
theta_df2 = pd.read_csv("../fit_IRT/IRTMatrixExtended_2PL_abilities.csv")

```

```python
# --- Percentage histogram (5% bins) ---
bins = np.arange(0, 105, 5)
plt.figure()
plt.hist(scores_df["Percentage"], bins=bins)
plt.xlabel("Percentage score")
plt.ylabel("Count of participants")
plt.title("Histogram of Percentage (5% bins)")
plt.xticks(bins, rotation=45)
plt.tight_layout()
plt.show()
```

```python
# --- Percentage histogram (5% bins) ---
bins = np.arange(0, 105, 5)
plt.figure()
plt.hist(scores_df2["Percentage"], bins=bins)
plt.xlabel("Percentage score")
plt.ylabel("Count of participants")
plt.title("Histogram of Percentage (5% bins)")
plt.xticks(bins, rotation=45)
plt.tight_layout()
plt.show()
```

```python
# --- Percentage histogram (5% bins) ---
bins = np.arange(0, 105, 5)
plt.figure()
plt.hist(scores_df2["Percentage"], bins=bins)
plt.xlabel("Percentage score")
plt.ylabel("Count of participants")
plt.title("Histogram of Percentage (5% bins)")
plt.xticks(bins, rotation=45)
plt.tight_layout()
plt.show()
```

```python
# --- Check Percentage = Score/21 * 100 ---
max_score = 21
scores_df["Percentage_from_formula"] = scores_df["Score"] / max_score * 100
scores_df["difference"] = scores_df["Percentage"] - scores_df["Percentage_from_formula"]
print(scores_df["difference"].describe())
```

```python
# --- Raw score distribution ---
plt.figure()
score_counts = scores_df["Score"].value_counts().sort_index()
plt.bar(score_counts.index, score_counts.values)
plt.xlabel("Raw score")
plt.ylabel("Count of participants")
plt.title("Distribution of raw scores")
plt.xticks(range(0, 22))
plt.tight_layout()
plt.show()
```

```python
# --- Normalised scores (z) ---
scores_df["Score_z"] = (scores_df["Score"] - scores_df["Score"].mean()) / scores_df["Score"].std()

plt.figure()
plt.hist(scores_df["Score_z"], bins=20)
plt.xlabel("Score (z)")
plt.ylabel("Count of participants")
plt.title("Distribution of normalised scores (z-scores)")
plt.tight_layout()
plt.show()
```

```python
# --- Theta histogram ---
plt.figure()
plt.hist(theta_df["theta"], bins=20)
plt.xlabel("Theta (ability)")
plt.ylabel("Count of participants")
plt.title("Distribution of IRT theta estimates")
plt.tight_layout()
plt.show()
```

```python
# --- Theta histogram ---
plt.figure()
plt.hist(theta_df2["theta"], bins=20)
plt.xlabel("Theta (ability)")
plt.ylabel("Count of participants")
plt.title("Distribution of IRT (extended) theta estimates")
plt.tight_layout()
plt.show()
```

```python


plt.figure()
plt.scatter(scores_df["Score"], theta_df["theta"], alpha=0.5)
plt.xlabel("Raw score")
plt.ylabel("Theta")
plt.title("Relationship between raw score and IRT ability (theta)")
plt.tight_layout()
plt.show()
```
